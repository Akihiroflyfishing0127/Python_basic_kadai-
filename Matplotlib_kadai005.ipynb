```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2,3, tight_layout=True)
#plt.show()

#折れ線グラフ
y0 = [1, 2, -5, 2]
x0 = [1,2,3,4]
axes[0,0].plot(x0,y0)


#SINグラフ
x1 = np.linspace(1,10,100)
y1=  2 + 2*np.sin(2*x1)

axes[0,1].plot(x1, y1, linewidth=2.0)
axes[0,1].set(xlim=(0,10), xticks=np.arange(0,10),
ylim=(-1,5), yticks=np.arange(-1,6))


#ヒストグラム 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x2 = np.random .normal(15, 5, 2000)
plt.xlim(-10, 40) #これが効いていない
# ヒストグラムをプロット
axes[0,2].hist(x2, bins=30)

#散布図
# データを生成する
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

# プロット
axes[1,0].scatter(x3, y3)


#図表示
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データ生成
x2 = np.random.normal(15, 5, 2000)

# 新しい図を作成
plt.figure()

# 1x1のサブプロットを作成
plt.subplot(2, 3, 1)
plt.hist(x2, bins=30)

# タイトルを設定
plt.title('Histogram of x2')

# 表示
plt.show()
```
