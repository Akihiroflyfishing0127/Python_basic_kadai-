```python
import requests
import pprint
from getpass import getpass

# APIキーを指定
api_key = getpass('APIキーを入力してください: ')

# 検索キーワードを変数search_wordに格納
search_word = input('駅名を入力してください: ')

# Google maps APIのURL
url = 'https://maps.googleapis.com/maps/api/geocode/json'

# データを取得
response = requests.get(
    url,
    params={
        'key': api_key,  # APIキーをパラメータに追加
        'address': search_word,  #駅名
    }
)

# レスポンスのJSONを取得
json_data = response.json()
# JSONの中身を確認
pprint.pprint(json_data)

geocode_data=response.json()
latitude=geocode_data['results'][0]['geometry']['location']['lat']
longitude=geocode_data['results'][0]['geometry']['location']['lng']


#レストラン情報のURL
urlR= 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

# データを取得
response = requests.get(
    urlR,
    params={
        'key': api_key,  # APIキーをパラメータに追加
        'location': f'{latitude},{longitude}',
        'language': 'Ja', #言語
        'radius': 500, #半径
        'type': 'restaurant',
    }
)

data = response.json()

#レストランの情報を抽出
restaurants=[]
for place in data['results']:
  restaurant={
      'name': place['name'],
      'rating': place.get('rating','N/A'),
      'address':place.get('vicinity','N/A'),
      'types': place.get('types',[]),
      'place_id': place['place_id']
  }
  restaurants.append(restaurant)

for restaurant in restaurants:
    print(restaurant['name'],restaurant['rating'],restaurant['address'])





```
